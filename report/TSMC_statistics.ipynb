```python
import pandas as pd
import matplotlib.pyplot as plt

# 讀取數據
tsmc_df = pd.read_csv('/Users/jennifer/Downloads/202001-202312_TSM.csv', parse_dates=['Date'])

# 計算基本統計量的函數
def calculate_statistics(df, column):
    return {
        'Mean': df[column].mean(),
        'Median': df[column].median(),
        'Standard Deviation': df[column].std(),
        'Max': df[column].max(),
        'Min': df[column].min(),
        '25th Percentile': df[column].quantile(0.25),
        '75th Percentile': df[column].quantile(0.75)
    }

# 計算台積電股價的基本統計量
tsmc_stats = calculate_statistics(tsmc_df, 'Close')

print("台積電股價基本統計量:")
for stat, value in tsmc_stats.items():
    print(f"{stat}: {value}")

# 繪製時間序列圖
plt.figure(figsize=(12, 6))
plt.plot(tsmc_df['Date'], tsmc_df['Close'], label='TSMC Stock Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('TSMC Stock Price Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
